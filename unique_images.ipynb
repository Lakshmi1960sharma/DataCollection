# Notebook to find duplicates or almost similar images and remove them from the folder 

```python
import os
import cv2
import imutils
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from os import listdir
from imaging_interview_exp import draw_color_mask,preprocess_image_change_detection,compare_frames_change_detection
```

```python
def find_duplicates(folder_dir,min_contour_area):
    """Finding the duplicates or almost similar images from a given path
    
    Parameters
    ----------
    folder_dir: 
        Location of the path of images in a folder
    min_contour_area: int
        Area of a contour to be skipped 
 
    Returns
    -------
    list_duplicate: list
        List of paths of images which looks identical or almost similar 
    """
    list_duplicate=[]
    for i,image in enumerate(os.listdir(folder_dir)):
        if (image.endswith(".png")):
            path=os.path.join(folder_dir,image)
            original_img= cv2.imread(path)
            print(f'original image {i} with {original_img.shape}')
            for j,compare in enumerate(os.listdir(folder_dir)):
                if (compare.endswith(".png")):
                    location=os.path.join(folder_dir,compare)
                    compared_image= cv2.imread(location)
                    if i<j and original_img.shape==compared_image.shape:
                        print(f'image {j} to compare with {compared_image.shape}')
                        previous_frame= preprocess_image_change_detection(original_img,[7,7])
                        next_frame= preprocess_image_change_detection(compared_image,[7,7])
                        score, res_cnts, thresh,frame_delta,cnts= compare_frames_change_detection(prev_frame=previous_frame, next_frame=next_frame, min_contour_area=min_contour_area)
                        print(f'score of image {i} compared with image {j} is {score} and length of wanted countours are {len(res_cnts)} from total available countours {len(cnts)}')
                        if score<5000:
                            list_duplicate.append(location)
    return np.unique(list_duplicate)
```

# How does your program work?

* Required libraries are imported
* I have used 'imaging_interview_exp.py' file which returns 'score', 'res_cnts', 'thresh','frame_delta','cnts' from 'compare_frames_change_detection function'.

* Find_duplicates function requires two parameters 
    1.Path of the folder directory
    2.Minimum contour area
* once entering into the function first image is compared from the second image to all the images in that folder and in the same way all the images are iterated
    *if both the image shapes are equal then those images are passed to the 'preprocess_image_change_detection' function
        *image passed to the 'preprocess_image_change_detection' does the following steps:
            *if no gaussian blur is applied to the image
                1. Converting an RGB image into a gray scale
                2. Resulting image is passed to the 'draw_color_mask' function which return the image with 5% of                              left boarder,5% of right border and 10% of top border is removed.
             *if gaussian blur is applied to the image:
                 1. Converting an RGB image into a gray scale
                 2. Gaussian blur is applied to the gray scale image and the kernel size is passes as a list [7,7]
                 2. Resulting image is passed to the 'draw_color_mask' function which return the image with 5% of                              left boarder,5% of right border and 10% of top border is removed.
                 
* Now, both the images are preprocessed and passed to the 'compare_frames_change_detection' function with the minimum contour area as 40. This function does the following steps:
     1. Absolute difference is computed over the two images.
     2. Threshold of 45 is applied to the resulting image where 0 is assigned to the pixel values less than 45 and 255 to           the pixel values greater than 45
     3. Dilation is applied to the image to whilting the areas.
     4. Contours are found over the resulting image
     5. If the area of a contour is less than the Minimum contour area then that particular contour is skipped.
     6. Remaining list of contour areas are computed and summed up and stored into the score function 
     7. If the score is less than 5000 then that image is categarized as similar, else dissimilar
     8. this function returns the location of similar images which is stored in similar_img_path
    
* for loop is iterated to similar_img_path and each similar image location is removed using os.remove
* In the end in the folder does only have unique images
